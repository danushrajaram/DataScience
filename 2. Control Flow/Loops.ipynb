```python
## while loop
count=1
while count<6:
  print(count)
  count+=1
```

```python
count=0
while count%2==0:
  print(count)
  count+=1
```

```python
## Loop Control Statements

# break

for i in range(10):
  if i ==5:
    break
  print(i)
```

```python
# Continue

# displaying odd numbers in range of 0-10
for i in range(10):
  if i%2==0:
    continue
  print(i)
```

```python
for i in range(15):
    if i % 3 == 0 and i != 0:  # Check if it's divisible by 3 (excluding 0)
        print(f"The number {i} is divisible by 3")
    else:
        print(i)  # Print other numbers normally
```

```python
# Nested loop

for i in range(3):
  for j in range(2):
    print(f"i:{i} and j:{j}")
```

```python
n=10
sum=0
count=1

while count<=n:
  sum=sum+count
  count=count+1

print(f"Sum of First N natural numbers {sum}")

```

```python
n=10
sum=0
count=1
for count in range(11):
  sum+=count
  count+=1
print(f"Sum of First N natural numbers {sum}")

```

```python

# Prime Numbers between 1 and 100

for num in range(1, 101):  # Include 100 in the range
    if num > 1:  # Prime numbers are greater than 1
        for i in range(2, num):
            if num % i == 0:  # If num is divisible by i, it's not prime
                break
        else:  # This else belongs to the for-loop, not the if statement!
            print(num)  # Print only if the loop completes without breaking
```
