**Question 5:** Write a Python program to calculate the factorial of a number.

Iterative Approach (Using for Loop)

```python
def factorial_iterative(n):
    fact = 1  # Step 1: Initialize fact to 1
    for i in range(1, n + 1):  # Step 2: Loop from 1 to n
        fact *= i  # Step 3: Multiply fact by current i
    return fact  # Step 4: Return the final factorial value

# Example usage
print(factorial_iterative(5))  # Output: 120
```

Recursive Approach

```python
def factorial_recursive(n):
    if n == 0 or n == 1:  # Step 1: Base case, return 1 if n is 0 or 1
        return 1
    return n * factorial_recursive(n - 1)  # Step 2: Recursive call

# Example usage
print(factorial_recursive(5))  # Output: 120



'''
factorial_recursive(5)
    → 5 * factorial_recursive(4)
        → 4 * factorial_recursive(3)
            → 3 * factorial_recursive(2)
                → 2 * factorial_recursive(1)
                    → returns 1 (Base Case)

                → 2 * 1 = 2  (Return to previous call)
            → 3 * 2 = 6  (Return to previous call)
        → 4 * 6 = 24  (Return to previous call)
    → 5 * 24 = 120  (Final Result)
'''
```

```python
def leapyear(n):
    if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
        print("The Year " + str(n) + " is a leap year")
    else:
        print("The Year " + str(n) + " is not a leap year")

# Test the function
leapyear(2024)  # Output: The Year 2024 is a leap year
leapyear(1900)  # Output: The Year 1900 is not a leap year
leapyear(2000)  # Output: The Year 2000 is a leap year
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Without Using a Function
numbers = [5, 2, 9, 1, 5, 6]
numbers.sort()  # Sorting the list in-place
print("Sorted list:", numbers)
```

```python
#Using a Function
def sort_numbers(num_list):
    return sorted(num_list)  # Using sorted() to return a new sorted list

numbers = [5, 2, 9, 1, 5, 6]
sorted_numbers = sort_numbers(numbers)
print("Sorted list:", sorted_numbers)
```

```python
#Without a Function (Using Bubble Sort)
numbers = [5, 2, 9, 1, 5, 6]

n = len(numbers)
for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:  # Swap if the element is greater than the next one
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print("Sorted list:", numbers)
```

```python
#With a Function (Using Selection Sort)

def selection_sort(num_list):
    n = len(num_list)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if num_list[j] < num_list[min_index]:  # Find the smallest element
                min_index = j
        num_list[i], num_list[min_index] = num_list[min_index], num_list[i]  # Swap
    return num_list

numbers = [5, 2, 9, 1, 5, 6]
sorted_numbers = selection_sort(numbers)
print("Sorted list:", sorted_numbers)
```
